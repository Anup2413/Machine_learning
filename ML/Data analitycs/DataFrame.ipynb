```python
#all detailsof mumbai indians when it won in hyderabad
#using multiple conditional filter, use logical operator
#use multiple conditions in python, use logical operator
'''
1. and  in python  
2. or  in python
3. not in python

but in pandas :
for and logical operator  then use & (ampercend)
for or logical operator then use | (pipeline)
for not logical operator then use ! (exclamination)
'''
#show record whoes winner is mumbai and venue is hyderabad
IPL[(IPL['Winner']=='Mumbai Indians') & (IPL['Venue']=='Hyderabad')] 
#while passing multiple condition allways pass in parenthisis
```

```python
#all detailsof mumbai indians when it won in hyderabad
#using multiple conditional filter, use logical operator
#use multiple conditions in python, use logical operator
'''
1. and  in python  
2. or  in python
3. not in python

but in pandas :
for and logical operator  then use & (ampercend)
for or logical operator then use | (pipeline)
for not logical operator then use ! (exclamination)
'''
#show record whoes winner is mumbai or venue is hyderabad
IPL[(IPL['Winner']=='Mumbai Indians') | (IPL['Venue']=='Hyderabad')] 
#while passing multiple condition allways pass in parenthisis
```
