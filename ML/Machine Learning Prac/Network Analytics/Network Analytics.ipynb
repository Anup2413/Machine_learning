```python
import networkx as nx
```

```python
#suppose we have a list of edge 
edgelist=[(1,2),(1,3),(1,4),(1,5),(1,6),(1,7),(1,8)] #list of tuple
print(edgelist)
```

```python
g=nx.Graph(edgelist)
#Graph() inbuilt method of networkx
#g is user defined object
```

```python
g
```

```python
#to draw graph
nx.draw(g)
#without lable
```

```python
#to draw with lable
nx.draw(g,with_labels=True)
```

```python
nx.draw_networkx(g) #another way to draw
```

```python
#to find out degree of node 1
#degree means direct conection
nx.degree(g,1)
```

```python
#to find out degree of node 2
nx.degree(g,2)
```

```python
#to find degree of all the nodes
nx.degree_centrality(g)
#degree_centrality's formula =
#degree_centrality of node 1(main node):
#=(degree of node 1)/total no. of nodes =7/7=1
#degree_centrality of node 2
#(degree of node 2)/total no. of nodes =1/7=0.14
'''
#degree_centrality of node 3
#(degree of node 3)/total no. of nodes =1/7=0.14
'''
```

```python
#to find the closness
#find closeness_centrality  : formula=1/total frequency of node
nx.closeness_centrality(g)
```

```python
#to find betweenness
nx.betweenness_centrality(g)
```
